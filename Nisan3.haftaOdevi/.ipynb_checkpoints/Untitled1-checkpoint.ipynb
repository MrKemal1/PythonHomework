```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("HouseDataClean.csv",index_col=0)#index_col=0 0. indeski bir index kolonuna dnüştürüp göstermez
```

```python
df.head()
```

```python
X= df.drop(columns=["MEDV"])
y=df["MEDV"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model=DecisionTreeRegressor(max_depth=4,random_state=42)
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae=mean_absolute_error(y_pred,y_test)
mae
```

```python
mse=mean_squared_error(y_pred,y_test)
mse
```

```python
r2=r2_score(y_pred,y_test)
r2

#Değer 1 e yakın ise sonuç iyidir -1 e yakın ise kötüdür
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20,10))
plot_tree(model, feature_names=X.columns, filled=True)
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={
    "max_depth":list(range(1,1000))
}
```

```python
grid = GridSearchCV(DecisionTreeRegressor(), param_grid,scoring="r2",cv=10)
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_,grid.best_params_,grid.best_score_
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2=RandomForestRegressor(n_estimators=10,random_state=42,max_depth=4)
model2.fit(X_train,y_train)
```

```python
y_pred2=model2.predict(X_test)
```

```python
mae2=mean_absolute_error(y_pred2,y_test)
mse2=mean_squared_error(y_pred2,y_test)
r2_2= r2_score(y_pred2,y_test)
```

```python
"Random Forest",mae2,mse2,r2_2
```
